```python
import numpy as np
import matplotlib.pyplot as plt

features = [
    "Severity",
    "Chronicity",
    "Voquezna Exp",
    "Specialist",
    "Comorbidity",
    "Polypharmacy",
    "Annual spend",
    "Treatment line"
]

baseline = np.array([3, 3, 3, 3, 3, 3, 3, 3])

prototypes = {
    "NI-7 NERD short-term": np.array([2, 2, 5, 1, 1, 1, 1, 3]),  
    
    "PP-5 Oncology GERD": np.array([3, 3, 1, 5, 5, 5, 5, 1]),  
    
    "NI-1B Barretts + EE": np.array([5, 5, 2, 5, 3, 3, 4, 5]),  
    
    "PP-2 GLP-1 GERD": np.array([3, 3, 3, 2, 4, 3, 3, 1]),  
    
    "PP-8 NSAID ulcers": np.array([3, 2, 3, 2, 3, 4, 3, 4]),  
    
    "PP-4 Antibiotics GERD": np.array([2, 1, 4, 1, 2, 3, 2, 2])  
}

def plot_force_like(name, values):
    contrib = values - baseline

    colors = ["#ff4d6d" if x > 0 else "#4da6ff" for x in contrib]

    plt.figure(figsize=(6, 2.2))

    plt.barh(features, contrib, color=colors)
    plt.axvline(0, color="black", linewidth=1)

    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    plt.tick_params(axis='y', length=0)
    plt.xlim(-2.5, 2.5)

    plt.title(name)
    plt.xlabel("Contribution vs baseline GERD patient")

    plt.tight_layout()
    plt.show()
```

```python
def jitter(vec, scale=0.55):
    return vec + np.random.normal(0, scale, size=len(vec))

prototypes_noisy = {k: jitter(v) for k, v in prototypes.items()}

for name, vec in prototypes_noisy.items():
    plot_force_like(name, vec)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_radar(name, values):
    values = np.clip(values, 0, 6) 

    categories = features
    N = len(categories)

    angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
    values = values.tolist()

    values += values[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(4.5, 4.5), subplot_kw=dict(polar=True))

    royal_blue = "#1f4e79"

    ax.plot(angles, values, linewidth=2.5, color=royal_blue)
    ax.fill(angles, values, alpha=0.5, color=royal_blue)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)

    ax.set_yticks([1, 2, 3, 4, 5, 6])
    ax.set_yticklabels([])

    # optional cleaner look
    #ax.grid(False)

    ax.set_title(name)

    plt.tight_layout()
    #plt.show()

    filename = name.replace(" ", "_").replace("-", "") + "_radar.svg"
    plt.savefig(filename, format="svg")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

np.random.seed(42)


n_gerd = 200
n_no_gerd = 800

# survival times (months on GLP-1)
# GERD  shorter persistence
time_gerd = np.random.exponential(scale=12, size=n_gerd)
time_no_gerd = np.random.exponential(scale=22, size=n_no_gerd)

# censor at 36 months
max_time = 36

event_gerd = (time_gerd <= max_time).astype(int)
event_no_gerd = (time_no_gerd <= max_time).astype(int)

time_gerd = np.minimum(time_gerd, max_time)
time_no_gerd = np.minimum(time_no_gerd, max_time)

df = pd.DataFrame({
    "time": np.concatenate([time_gerd, time_no_gerd]),
    "event": np.concatenate([event_gerd, event_no_gerd]),
    "group": ["GLP1 + GERD"] * n_gerd + ["GLP1 no GERD"] * n_no_gerd
})

kmf_gerd = KaplanMeierFitter()
kmf_no = KaplanMeierFitter()

plt.figure(figsize=(5, 5))

kmf_gerd.fit(
    df[df["group"] == "GLP1 + GERD"]["time"],
    df[df["group"] == "GLP1 + GERD"]["event"],
    label="GLP-1 + GERD"
)
kmf_gerd.plot(ci_show=True, linewidth=2, color="#2C888B")

kmf_no.fit(
    df[df["group"] == "GLP1 no GERD"]["time"],
    df[df["group"] == "GLP1 no GERD"]["event"],
    label="GLP-1 no GERD"
)
kmf_no.plot(ci_show=True, linewidth=2, color="#461B62")

plt.title("Kaplan–Meier: GLP-1 Persistence ~ GERD Status")
plt.xlabel("\n Months on GLP-1")
plt.ylabel("Persistence probability \n")
plt.xlim(0, 36)
plt.ylim(0, 1.05)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)

plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams.update({
    "font.weight": "bold",
    "axes.labelweight": "bold",
    "axes.titleweight": "bold",
})

ages = list(range(18, 91))
n_ages = len(ages)

population_male = []
population_female = []

for age in ages:
    if age < 30:
        male_pct = 0.5
        female_pct = 0.8
    elif age < 45:
        male_pct = 1.2 + (age - 30) * 0.1
        female_pct = 1.8 + (age - 30) * 0.12
    elif age < 65:
        male_pct = 2.5 + (age - 45) * 0.15
        female_pct = 3.5 + (age - 45) * 0.18
    elif age < 75:
        male_pct = 5.0 + (age - 65) * 0.1
        female_pct = 6.5 + (age - 65) * 0.1
    else:
        male_pct = 6.0 - (age - 75) * 0.08
        female_pct = 7.5 - (age - 75) * 0.1
    
    population_male.append(max(0.3, male_pct))
    population_female.append(max(0.4, female_pct))

total_male = sum(population_male)
total_female = sum(population_female)
total_all = total_male + total_female

population_male_pct = [p/total_all*100 for p in population_male]
population_female_pct = [p/total_all*100 for p in population_female]

fig, ax = plt.subplots(figsize=(8, 5))

width = 0.35
x_pos = ages

x_male = [x - width/2 for x in x_pos]
x_female = [x + width/2 for x in x_pos]

ax.bar(x_male, population_male_pct, width, 
       label='Male (population)', color="#0d3b69", alpha=0.7, 
       edgecolor='black', linewidth=0.5)

ax.bar(x_female, population_female_pct, width, 
       label='Female (population)', color="#af364c", alpha=0.7, 
       edgecolor='black', linewidth=0.5)

ax.set_ylabel('Proportion (%)', fontsize=11)
ax.set_xlabel('Age (years)', fontsize=11)
ax.set_title('Target population (real world) by age and gender', fontsize=13, fontweight='bold')

ax.set_xlim(17, 91)
ax.set_xticks(ages[::5])
ax.set_xticklabels(ages[::5], rotation=45, ha='right')

ax.legend(loc='upper right')

ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams.update({
    "font.weight": "bold",
    "axes.labelweight": "bold",
    "axes.titleweight": "bold",
})

ages = list(range(18, 91))
n_ages = len(ages)



np.random.seed(42)  

sample_male = []
sample_female = []

for age in ages:
    if age < 35:
        # Young: overrepresented (peak at 25-30)
        base_male = 8 * (1 - abs(age - 28)/20) + 2
        base_female = 6 * (1 - abs(age - 26)/20) + 1
    elif age < 55:
        # Middle: declining representation
        base_male = 5 * (1 - (age - 35)/25) + 1
        base_female = 4 * (1 - (age - 35)/25) + 0.5
    else:
        # Elderly: severely underrepresented
        base_male = 0.5 * (1 - (age - 55)/35) + 0.05
        base_female = 0.3 * (1 - (age - 55)/35) + 0.02
    
    # Add small random noise to sample (looks realistic)
    noise_male = np.random.normal(0, base_male * 0.05)
    noise_female = np.random.normal(0, base_female * 0.05)
    
    sample_male.append(max(0.05, base_male + noise_male))
    sample_female.append(max(0.02, base_female + noise_female))

total_sample = sum(sample_male) + sum(sample_female)
sample_male_pct = [s/total_sample*100 for s in sample_male]
sample_female_pct = [s/total_sample*100 for s in sample_female]



population_male = []
population_female = []

for age in ages:
    # Using a smooth, realistic epidemiological curve (log-normal shaped)
    # Prevalence increases slowly in young, peaks at 65-70, then declines slightly
    
    # Rescale age to 0-1 range for smooth function
    t = (age - 18) / 72  # 0 at 18, 1 at 90
    
    # Asymmetric curve: slow rise, peak around 0.7 (age ~68), gentle decline
    # Using beta distribution shape
    male_curve = 5.5 * (t**1.2) * ((1-t)**0.6) + 0.5
    female_curve = 7.0 * (t**1.1) * ((1-t)**0.5) + 0.8
    
    # Add tiny noise for realism (less noise than sample)
    male_val = male_curve + np.random.normal(0, male_curve * 0.02)
    female_val = female_curve + np.random.normal(0, female_curve * 0.02)
    
    population_male.append(max(0.3, male_val))
    population_female.append(max(0.4, female_val))






total_pop = sum(population_male) + sum(population_female)
population_male_pct = [p/total_pop*100 for p in population_male]
population_female_pct = [p/total_pop*100 for p in population_female]





weights_male = []
weights_female = []

for i in range(n_ages):
    w_m = population_male_pct[i] / sample_male_pct[i] if sample_male_pct[i] > 0 else 0
    w_f = population_female_pct[i] / sample_female_pct[i] if sample_female_pct[i] > 0 else 0
    weights_male.append(min(w_m, 12))
    weights_female.append(min(w_f, 12))






fig, axes = plt.subplots(3, 1, figsize=(8, 12))

width = 0.35
x_pos = ages
x_male = [x - width/2 for x in x_pos]
x_female = [x + width/2 for x in x_pos]

# Panel 1: Sample
ax1 = axes[0]
ax1.bar(x_male, sample_male_pct, width, label='Male (sample)', 
        color='#0d3b69', alpha=0.7, edgecolor='black', linewidth=0.5)
ax1.bar(x_female, sample_female_pct, width, label='Female (sample)', 
        color='#af364c', alpha=0.7, edgecolor='black', linewidth=0.5)
ax1.set_ylabel('Proportion (%)', fontsize=11)
ax1.set_title('Observed sample (raw data) - biased: young overrepresented, elderly underrepresented', fontsize=12)
ax1.set_xlim(17, 91)
ax1.set_xticks(ages[::10])
ax1.set_xticklabels(ages[::10])
ax1.legend(loc='upper right')
ax1.grid(axis='y', linestyle='--', alpha=0.3)

# Panel 2: Population
ax2 = axes[1]
ax2.bar(x_male, population_male_pct, width, label='Male (population)', 
        color='#0d3b69', alpha=0.7, edgecolor='black', linewidth=0.5)
ax2.bar(x_female, population_female_pct, width, label='Female (population)', 
        color='#af364c', alpha=0.7, edgecolor='black', linewidth=0.5)
ax2.set_ylabel('Proportion (%)', fontsize=11)
ax2.set_title('Target population (real world) - natural curve: low in young, peaks at 65-70, slightly higher for females', fontsize=12)
ax2.set_xlim(17, 91)
ax2.set_xticks(ages[::10])
ax2.set_xticklabels(ages[::10])
ax2.legend(loc='upper right')
ax2.grid(axis='y', linestyle='--', alpha=0.3)

# Panel 3: Weights
ax3 = axes[2]
ax3.plot(ages, weights_male, linewidth=2, marker='', label='Weight (male)', color='#0d3b69')
ax3.plot(ages, weights_female, linewidth=2, marker='', label='Weight (female)', color='#af364c')
ax3.axhline(y=1, linestyle='--', color='gray', alpha=0.7, linewidth=1.5, label='Weight = 1 (neutral)')
ax3.fill_between(ages, 1, weights_male, where=(np.array(weights_male) > 1), alpha=0.2, color='#0d3b69')
ax3.fill_between(ages, 1, weights_female, where=(np.array(weights_female) > 1), alpha=0.2, color='#af364c')
ax3.set_ylabel('Calibration weight', fontsize=11)
ax3.set_xlabel('Age (years)', fontsize=11)
ax3.set_title('Calibration weights: Higher weight = underrepresented in sample', fontsize=12)
ax3.set_xlim(17, 91)
ax3.set_ylim(0, 13)
ax3.set_xticks(ages[::10])
ax3.set_xticklabels(ages[::10])
ax3.legend(loc='upper right')
ax3.grid(axis='both', linestyle='--', alpha=0.3)



plt.suptitle('From raw data to decision-ready: Calibration by age and gender', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
